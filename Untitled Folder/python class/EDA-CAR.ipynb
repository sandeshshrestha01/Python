```python
# univariate anaysis of num_cols
for col in num_cols:
    print(col)
    plt.figure(figsize= (15,4))
    plt. subplot(1,2,1)
    df[col].hist()
    plt.ylabel('count')
    plt.subplot(1,2,2)
    plt.boxplot(x=df[col])
    plt.show()
    
```

```python
# Categorical variables
fig, axes= plt.subplots(3,2, figsize=(18,18))
sns. countplot(ax=axes[0,0], x= 'Fuel_Type', data = df , order= df ['Fuel_Type'].value_counts().index)
sns. countplot(ax=axes[0,1], x= 'Transmission', data = df , order= df ['Transmission'].value_counts().index)
sns. countplot(ax=axes[1,0], x= 'Owner_Type', data = df , order= df ['Owner_Type'].value_counts().index)
sns. countplot(ax=axes[1,1], x= 'Brand', data = df , order= df ['Brand'].head(10).value_counts().index)
sns. countplot(ax=axes[2,0], x= 'Model', data = df , order= df ['Model'].head(10).value_counts().index)
sns. countplot(ax=axes[2,1], x= 'Location', data = df , order= df ['Location'].value_counts().index)
```
