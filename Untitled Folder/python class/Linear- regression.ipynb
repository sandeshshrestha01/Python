## Selectinf Targate and features

```python
# our targetvariable is price
y= df.Price

features=['Rooms','Distance','Bedroom2', 'Bathroom','Car','Landsize','YearBuilt']
```

```python
df.columns
```

```python
# features variable 
X = df[features]
```

```python
X.corr()
```

### Build Models

```python
# import from sciki_learn
from sklearn.tree import DecisionTreeRegressor
```

```python
# Build a model
house_model = DecisionTreeRegressor(random_state=1)
```

```python
# fil the data the model
house_model.fit(X,y)
```

```python
X.head()
```

```python
# make prediction
house_model.predict(X.head())
```

```python
df['Price'].head()
```

```python
# Using train test split methord
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_X,test_X,train_y, test_y = train_test_split(X,y,random_state =0)
```

```python
house_model_new = DecisionTreeRegressor()
```

```python
house_model_new.fit(train_X, train_y)
```

```python
house_model_new.predict(train_X.head())
```

```python
#predict on test data
house_model_new=house_model_new.predict(test_X)
```

```python
test_y.head()
```

```python
# evaluate model performance 
#MAE (mean Absulute Error)

from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(test_y ,house_model_new)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(test_y, house_model_new)
```

```python
from sklearn.metrics import root_mean_squared_error
```

```python
root_mean_squared_error(test_y, house_model_new)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(test_y, house_model_new)
```
