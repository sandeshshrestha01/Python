```python
import matplotlib.pyplot as plt
```

```python
months= ['Jan','Feb','Mar','Apr','May','Jun','Jul','Agu','Sep','Oct','Nov','Dec']
sales=[5000,7000,8000,6000,11000,10500,12000,13000,15000,16000,21000,12000]
#creatw a line plot
plt.figure(figsize=(10,6))
plt.plot(months,sales )# line chart
#adding times and labels
plt.title("Monthly Sales figure")
plt.xlabel("months")
plt.ylabel("Salesin $")
plt.grid()
#show plot
plt.show()
```

```python
import pandas as pd 
import numpy as np 
import plotly.express as px
data ={
    'months':['Jan','Feb','Mar','Apr','May','Jun','Jul','Agu','Sep','Oct','Nov','Dec'],
    'sales':[5000,7000,8000,6000,11000,10500,12000,13000,15000,16000,21000,12000]
}
df= pd.DataFrame(data)
fig= px.line(df,x='months',y='sales', title='Months_sales',markers='o')
fig.show()
```

```python
# area plot
# sample data 
months =['Jan','Feb','Mar','Apr','May','Jun']
product_a = [30,50,60,40,45,70]
product_b = [70,45,40,60,50,30]
product_c = [40,60,45,30,70,50]

plt. figure(figsize= (10,6))
plt.fill_between(months, product_a, color='skyblue', alpha = 0.6 ,label='product_a ')
plt.fill_between(months, product_b, color='green', alpha = 0.6 ,label='product_b')
plt.fill_between(months, product_c, color='red', alpha = 0.6 ,label='product_c')
plt.grid(axis='y')
plt.show()
```

```python
# correlation matrix
data={
    'A':[1,2,3,4,5],
    'B':[5,4,3,2,1],
    'C':[1,3,5,2,4],
    'D':[4,2,1,3,5]
}
corr_df=pd.DataFrame(data)
corr_matrix= corr_df.corr()
corr_matrix 
```
